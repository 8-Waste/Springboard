Facial recognition requires a sequence of related steps including locating faces in an image, centering and measuring unique facial features, comparing to known faces and making a prediction.<br><br>
The data source for this project is photographs of fútbol players.  The photographs are primarily in a proprietary Canon format.  After consolidating the photos, they are converted, if necessary, from the proprietary Canon format (.CR2) into TIFF files so we can easily access them in python.<br>

We then train a model that identifies sharp vs blurry photos.  This model is run against the photos we will be using to detect faces and eliminate blurry photos.<br>

To identify a player, we need at minimum 30 photos of that player.  To help achieve this, we run a face detection algorism against our photos and save the faces in a separate location.  We can then browse that file and segregate photos of specific players.

This jupyter notebook includes a sample of the photos with players, training and applying a blur detection algorithm to the photos and my initial attempts at identifying faces utilizing a Harr-cascade classifier and a MTCNN classifier.

Train the model...

```python
# split data, train and save model
@time_it
def run_train():
    def f_importances(coef, names, top=-1):
        imp = coef
        imp, names = zip(*sorted(list(zip(imp, names))))

        # Show all features
        if top == -1:
            top = len(names)

        plt.barh(range(top), imp[::-1][0:top], align='center')
        plt.title('Feature Importance')
        plt.yticks(range(top), names[::-1][0:top])
        plt.show()

    images=pd.DataFrame()

    images = images.append(sharp_df)
    images = images.append(defocused_df)
    images = images.append(motion_df)

    feature_names = ['laplace-mean','laplace-var','laplace-max',
                     'sobel-mean','sobel-var','sobel-max',
                     'roberts-mean','roberts-var','roberts-max',
                     'canny-mean','canny-var','canny-max']
    images.columns = feature_names

    X = np.array(images)
    y = np.concatenate((np.ones((sharp_df.shape[0], )), np.zeros((defocused_df.shape[0]+motion_df.shape[0], ))), axis=0)

    X = scale(X)

    x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, stratify=y)

    svm_model = svm.SVC(C=125, kernel='linear', gamma='scale')
    svm_model.fit(x_train, y_train)
    pred =svm_model.predict(x_valid)
    print(f'Accuracy: {accuracy_score(y_valid, pred):.2%}')
    print(f'F1_score: {f1_score(y_valid, pred):.2%}\n')
    print('Confusion matrix:\n',confusion_matrix(y_valid, pred))

    print('\nClassification_report:\n',classification_report(y_valid, pred))

    f_importances(abs(svm_model.coef_[0]), images.columns, top=-1)
    return svm_model

svm_model = run_train()
```

Find faces (harr-cascade)...

```python
def find_harr_faces():
    import os
    import cv2
    row_list=[]
    files = df_s.query('sharp==1')['path_file']
    for c, file in enumerate(files):
        print(str(c+1) + ' of ' + str(len(files)), end='\r')
        Path(file.parent / 'harr_faces').mkdir(parents=True, exist_ok=True)
        Path(file.parent / 'harr_no_faces').mkdir(parents=True, exist_ok=True)
        image = cv2.imread(str(file))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
        faces = faceCascade.detectMultiScale(
            gray,
            scaleFactor=1.3,
            minNeighbors=3,
            minSize=(60, 60)
        )
        for i, (x, y, w, h) in enumerate(faces):
            face_file = file.parent / 'harr_faces' / (file.stem + '_faceno_' + str(i).zfill(4)+'.jpg')
            no_face_file = file.parent / 'harr_no_faces' / (file.stem + '_faceno_' + str(i).zfill(4)+'.jpg')
            dict = {'face_file':face_file, 'no_face_file': no_face_file, 'path_file':file, 'faceno':str(i).zfill(4), 'faceno_int':i, 'x1':x, 'y1':y, 'x2': x+w, 'y2':y+h, 'w':w, 'h':h, 'sq_pixels': w*h, 'org_image_h': image.shape[0], 'org_image_w':image.shape[1], 'org_image_sq_pixels':image.shape[0] * image.shape[1], 'face_pct': (w*h)/(image.shape[0] * image.shape[1])}
            roi_face_color = image[y:y + h, x:x + w]
            x1 = random.randint(0, image.shape[1]-w-1)
            y1 = random.randint(0, image.shape[0]-h-1)
            roi_noface_color = image[y1:y1 + h, x1:x1 + w]
            row_list.append(dict)
            cv2.imwrite(str(face_file), roi_face_color)
            cv2.imwrite(str(no_face_file), roi_noface_color)
    return pd.DataFrame(row_list)

# face detection using Harr-cascade classifier
df_harr = find_harr_faces()
```

Find faces (mtcnn)

```python
# utility function for mtcnn detecting faces
from mtcnn import MTCNN
def find_mtcnn_faces():
    import os
    import cv2
    row_list=[]
    files = df_s.query('sharp==1')['path_file']
    for c, file in enumerate(files):
        print(str(c+1) + ' of ' + str(len(files)) + '  '+ file.name + '                    ', end='\r')
        Path(file.parent / 'mtcnn_faces').mkdir(parents=True, exist_ok=True)
        Path(file.parent / 'mtcnn_no_faces').mkdir(parents=True, exist_ok=True)
        detector = MTCNN()  # we are finding this on the scaled image and storing points based on the scaled image
        image = cv2.imread(str(file))
        result = detector.detect_faces(image)
        for i, face in enumerate(result):
            (x, y, w, h) = face['box']
            confidence = face['confidence']
#             key_points = face['keypoints']
            org_image_left_eye = face['keypoints']['left_eye']
            org_image_right_eye = face['keypoints']['right_eye']
            org_image_nose = face['keypoints']['nose']
            org_image_mouth_left = face['keypoints']['mouth_left']
            org_image_mouth_right = face['keypoints']['mouth_right']
            face_image_left_eye = tuple(np.subtract(org_image_left_eye, (x, y)))
            face_image_right_eye = tuple(np.subtract(org_image_right_eye, (x, y)))
            face_image_nose = tuple(np.subtract(org_image_nose, (x, y)))
            face_image_mouth_left = tuple(np.subtract(org_image_mouth_left, (x, y)))
            face_image_mouth_right = tuple(np.subtract(org_image_mouth_right, (x, y)))
            
            face_file = file.parent / 'mtcnn_faces' / (file.stem + '_faceno_' + str(i).zfill(4)+'.jpg')
            no_face_file = file.parent / 'mtcnn_no_faces' / (file.stem + '_faceno_' + str(i).zfill(4)+'.jpg')
            dict = {'face_file':face_file, 'no_face_file': no_face_file, 'path_file':file, 'faceno':str(i).zfill(4), 'faceno_int':i, 'x1':x, 'y1':y, 'x2': x+w, 'y2':y+h, 'w':w, 'h':h, 'confidence':confidence, 'sq_pixels': w*h,  'org_image_h': image.shape[0], 'org_image_w':image.shape[1], 'org_image_sq_pixels':image.shape[0] * image.shape[1], 'face_pct': (w*h)/(image.shape[0] * image.shape[1]), 'org_image_left_eye':org_image_left_eye, 'org_image_right_eye':org_image_right_eye, 'org_image_nose':org_image_nose, 'org_image_mouth_left':org_image_mouth_left, 'org_image_mouth_right':org_image_mouth_right, 'face_image_left_eye':face_image_left_eye, 'face_image_right_eye':face_image_right_eye, 'face_image_nose':face_image_nose, 'face_image_mouth_left':face_image_mouth_left, 'face_image_mouth_right':face_image_mouth_right}
            roi_face_color = image[y:y + h, x:x + w]
            x1 = random.randint(0, image.shape[1]-w-1)
            y1 = random.randint(0, image.shape[0]-h-1)
            roi_noface_color = image[y1:y1 + h, x1:x1 + w]
            row_list.append(dict)
            cv2.imwrite(str(face_file), roi_face_color)
            cv2.imwrite(str(no_face_file), roi_noface_color)
    return pd.DataFrame(row_list)

# face detection using Harr-cascade classifier
df_mtcnn = find_mtcnn_faces()
```
